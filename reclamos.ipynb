# Feature Selection

```python
df = pd.read_csv('join.csv')
df
```

```python
df.select_dtypes(['number']).corr().abs().sort_values('ValorReclamo', ascending=False)[['ValorReclamo', 'Valuacion_Danios']]
```

```python
recvars = ['daniossuprec', 'reclamosup', 'recmej']
features = [col for col in df.select_dtypes(['number']).columns if 'rec' not in col.lower()]
features
```

# Learning

```python
from sklearn import model_selection, ensemble, tree, metrics
```

```python
datos = df.sample(frac=1).reset_index(drop=True) # Permuto las filas al azar x si estaban ordenadas
y = datos['ValorReclamo']
X = datos.loc[:, features]

y.describe()
```

```python
X.describe()
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y)
```

```python
arbol = tree.DecisionTreeRegressor()
arbol.fit(X_train, y_train)
```
