```python
import geopandas as gpd
import pandas as pd
```

# Parcelas

```python
sanroque = gpd.read_file('parcelas/sanroque/parcelas.shp')
santiago = gpd.read_file('parcelas/santiago/parcelas.shp')
parcelas = pd.concat([sanroque, santiago])

parcelas
```

```python
parcelas['centroid'] = parcelas.centroid
parcelas['area'] = parcelas.area
parcelas.set_geometry('centroid')
```

```python
parcelas.plot()
```

# Reclamos

```python
claims = pd.read_csv('Claims-WorkingFile_v2.csv')
claims['idParcela'] = claims['idParcela'].apply(lambda id: int(id.split('-')[0]))
claims[claims['Validado'] == 'X']
```

```python
claims.describe()
```

# Join

```python
join = parcelas.set_index('par_idparc').join(claims.set_index('idParcela'), how='right')
join.to_csv('join.csv')
```

```python
join.describe()
```

# Test

```python
nona = join['Valuacion_Danios'].notna()
sina = join['Valuacion_Danios'].isna()
```

```python
join['Valuacion'] = join['Valuacion'].fillna(join['Valuacion'].mean())
```

```python
linear = LinearRegression(fit_intercept=False)
linear.fit(pd.DataFrame(join['Valuacion'][nona]), join['Valuacion_Danios'][nona])
```

```python
join['Valuacion_Danios'][sina] = linear.predict(pd.DataFrame(join['Valuacion'][sina]))
```
