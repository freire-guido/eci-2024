```python
import json
import pandas as pd
```

# Parcelas

```python
with open('parcelas/santiago_parcelas.json') as f:
    santiago = json.load(f)
with open('parcelas/sanroque_parcelas.json') as f:
    sanroque = json.load(f)

santiago['features']
```

```python
santiago = pd.DataFrame([{**parcela['properties'], **parcela['geometry']} for parcela in santiago['features']])
santiago.to_csv('santiago.csv')
```

```python
santiago.describe()
```

```python
sanroque = pd.DataFrame([{**parcela['properties'], **parcela['geometry']} if parcela['geometry'] else parcela['properties'] for parcela in sanroque['features']])
sanroque.to_csv('sanroque.csv')
```

```python
parcelas = pd.concat([santiago, sanroque])
parcelas
```

# Reclamos

```python
claims = pd.read_csv('Claims-WorkingFile_v2.csv')
claims['idParcela'] = claims['idParcela'].apply(lambda id: int(id.split('-')[0]))
claims[claims['Validado'] == 'X']
```

```python
claims.describe()
```

# Join

```python
join = claims.set_index('idParcela').join(parcelas.set_index('par_idparcela'))
join.to_csv('join.csv')
```

```python
join.describe()
```

# Analisis

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
join
```

```python
def coord2np(coord):
    try:
        return np.squeeze(np.array(coord))
    except Exception:
        return coord
    
def arr2centroide(arr):
    try:
        return np.mean(arr, axis=0)
    except Exception:
        return arr
    
def getcoord(centroide, coord):
    try:
        return float(centroide[coord])
    except Exception:
        return pd.NA
```

```python
join['coordinates'] = join['coordinates'].apply(coord2np)
centroide = join['coordinates'].apply(arr2centroide)
centroide
```

```python
join['x'] = centroide.apply(lambda c: getcoord(c, 0))
join['y'] = centroide.apply(lambda c: getcoord(c, 1))
join[['x', 'y']]
```

```python
join.loc[join['x'].notna()].plot.scatter('x', 'y', c='ValorReclamo')
```

```python
join.loc[join['x'].notna()].plot.scatter('x', 'y', c='Valuacion_Danios')
```

```python
join.select_dtypes(['number']).corr()[['Valuacion_Danios', 'ValorReclamo']].sort_values(['Valuacion_Danios', 'ValorReclamo'], ascending=False)
```

```python
features = ['Valuacion', 'Valuacion_Mejoras', 'Superficie_Mejoras', 'ValorReclamo']
for feat in features:
    join.plot.scatter(feat, 'Valuacion_Danios')
```

```python
features
```

```python
nona = join[features + ['Valuacion_Danios']].notna().all(axis=1) # Entreno sin NA
X = join[features][nona]
y = join['Valuacion_Danios'][nona]

linear = LinearRegression()
linear.fit(X, y)
```

# Test

```python
test = pd.read_csv('test.csv')
X_test = join.loc[test['idParcela'].apply(lambda id: int(id.split('-')[0])), features]
X_test = X_test.fillna(X_test.mean()) # Predigo con NA = promedio
```

```python
X_test
```

```python
y_test = linear.predict(X_test)
y_test.clip(0, X_test['ValorReclamo'])
```

```python
test['pred'] = y_test
test.to_csv('pred.csv')
```
