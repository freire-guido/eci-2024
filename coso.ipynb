```python
import geopandas as gpd
import pandas as pd
```

# Parcelas

```python
sanroque = gpd.read_file('parcelas/sanroque.shp')
santiago = gpd.read_file('parcelas/santiago.shp')
parcelas = pd.concat([sanroque, santiago])

parcelas
```

```python
parcelas['centroid'] = parcelas.centroid
parcelas.set_geometry('centroid')
```

```python
parcelas.plot()
```

# Reclamos

```python
claims = pd.read_csv('Claims-WorkingFile_v2.csv')
claims['idParcela'] = claims['idParcela'].apply(lambda id: int(id.split('-')[0]))
claims[claims['Validado'] == 'X']
```

```python
claims.describe()
```

# Join

```python
join = parcelas.set_index('par_idparc').join(claims.set_index('idParcela'))
join.to_csv('join.csv')
```

```python
join.describe()
```

# Analisis

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
join.columns
```

```python
join.select_dtypes(['number']).corr()[['Valuacion_Danios', 'ValorReclamo']].sort_values(['Valuacion_Danios', 'ValorReclamo'], ascending=False)
```

```python
join.plot('Valuacion', legend=True)
```

```python
features = ['Valuacion', 'Valuacion_Mejoras', 'Superficie_Mejoras', 'ValorReclamo']
for feat in features:
    join.plot.scatter(feat, 'Valuacion_Danios')
```

```python
features
```

```python
nona = join[features + ['Valuacion_Danios']].notna().all(axis=1) # Entreno sin NA
X = join[features][nona]
y = join['Valuacion_Danios'][nona]

linear = LinearRegression()
linear.fit(X, y)
```

# Test

```python
test = pd.read_csv('test.csv')
X_test = join.loc[test['idParcela'].apply(lambda id: int(id.split('-')[0])), features]
X_test = X_test.fillna(X_test.mean()) # Predigo con NA = promedio
```

```python
X_test
```

```python
y_test = linear.predict(X_test)
test['Valuacion_Danios'] = y_test.clip(X_test['ValorReclamo'].mean(), np.array(X_test['ValorReclamo']))
test.to_csv('pred.csv', index=False)
```
